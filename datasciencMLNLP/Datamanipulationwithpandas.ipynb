```python
## renaming columns

df=df.rename(columns={'sale date':'sales date'})
df
```

```python
## change datatypes
df['Value_new']=df['Index'].astype(float)
df.head()
```

```python
df['New Value']=df['Index'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating and Grouping
df.head()
```

```python
grouped_mena=df.groupby('Country')['Index'].mean()
print(grouped_mena)
```

```python
df.groupby(['Country','City'])['Index'].sum()
```

```python
## aggregrate multiple function
grouped_agg=df.groupby('City')['Index'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
##Merging and Joining dataframes
# create samle dataframe

df1=pd.DataFrame({'key':['A','B','C'], 'Value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'], 'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge DataFrame on the 'Key Column'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
