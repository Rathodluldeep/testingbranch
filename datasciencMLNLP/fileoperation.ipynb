```python
## read a whole file
with open('example.txt','r')as file:
    content=file.read()
    print(content)
```

```python
## read a file line by line

with open('example.txt','r')as file:
    for line in file:
        print(line)
```

```python
with open('example.txt','r')as file:
    for line in file:
        print(line.strip()) ## strip() removes the new line
```

```python
## writing a file(overwriting)

with open('example.txt','w')as file:
    file.write("hello world!")
    file.write("this is a new line")
```

```python
## append the file 

with open('example.txt','w')as file:
    file.write("Append operation taking place!\n")
    
```

```python
## writing a list of line to a file

lines=['first line\n','Second line \n','Third line\n']
with open('example.txt','a')as file:
    file.writelines(lines)
```

```python
### Binary files

# writing to a binary file
data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb')as file:
    file.write(data)
```

```python
 ## read the content from a source file and write to a destiantion text file
with open('example.txt','r') as source_file:
     content=source_file.read()
    
with open('destination.txt','w')as destination_file:
    destination_file.write(content)
```

```python
#read a text file count the number of lines,words,and characters.
#Counting lines,word,and charaters in a text file

def count_text_file(file_path):
    with open(file_path,'r')as file:
        lines=file.readlines()
        line_count=(len(lines))
        word_count=sum(len(line.split()) for line in lines)
        char_count=sum(len(line) for line in lines)
    return line_count,word_count,char_count


file_path='example.txt'
lines,words,characters=count_text_file(file_path)
print(f'line:{lines},words:{words},character:{characters}')
```
