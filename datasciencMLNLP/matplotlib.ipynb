Data Visualization with Matplotlib
Matplotlib is a powerful plotting library for python that enable the creation of static,animated,and interactive visualizations.It is widely used for data visualization in data science and analysis. In this lesson,we will cover the basics of matplotlib,including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

##create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
##create a customized line plot
plt.plot(x,y,color='red',linestyle='--',marker='*',linewidth=3,markersize='9')
plt.grid(True)
```
