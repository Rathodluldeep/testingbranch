Memory Management Best Practices <br>
1.Use local variables:-Local variables have a shorter lifespan and are freed sooner than global variables.<br>
2.Avoid circular refrences:-Circular refrence can lead to memory leaks if not porperly managerd.<br>
3.Use Generators:Generators produce item one at a time and only keep one item in memory at a time,making them memory efficient.<br>
4.Explicitly Delete Objects:Use the del statement to delete variables and objects explicitly.<br>
5.Profile Memory Usage:-Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.<br>

```python
class MyObject:
    def __init__(self, name):
        self.name = name
        self.ref = None
    def __del__(self):
        print(f"{self.name} is being garbage collected")
    
    
##CReate circular refrence
obj1 = MyObject("Object1")
obj2 = MyObject("Object2")
obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2

##Manually trigger the garbage collector
gc.collect()
```

```python
## Generators for memory efficiency
## Generators allows you to produce item one at a time, which can be more memory efficient than creating a list of all items at once.

def genrate_numbers(n):
    for i in range(n):
        yield i
    
# using the generator
for num in genrate_numbers(100000):
    print(num)
    if num > 10:
        break
```

```python
##Profiling memory usage with tracemalloc
import tracemalloc

def create_list(n):
    return [i for i in range(n)]

def main():
    tracemalloc.start()
    
    create_list(100000)
    
    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')
    
    print("Top 10 lines")
    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
