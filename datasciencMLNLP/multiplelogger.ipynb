Logging with Multiple Loggers
You create multiple logger for diffrent part of your application

```python
import logging
##create a logger from modulel
logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

##CRetae alogger for module2
logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)
                          
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    force=True,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
##log message with diffrent loggers
logger1.debug("This is debug message for module1 ")
logger2.warning("This is a warning message for module 2")
logger2.error("This is an error message")
```
