```python
##instance variable and methods
class Dog:
    ## constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
## create object
dog1=Dog("buddy",3)
print(dog1)
print(dog1.name)
print(dog1.age)

    
```

```python
dog2=Dog("Lucy",5)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
##define a class with instance mwthods
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def bark(self):
        print(f"{self.name} says woof")

dog1=Dog("busy",3)
dog1.bark()
```

```python
## modeling a bank account
## define the class of bank acount
class BankAccount:
    def __init__(self,owner,balace=0):
        self.owner=owner
        self.balance=balace

    def deposite(self,amount):
        self.balance += amount
        print(f"{amount} is deposite .New balacne is {self.balance}")
    
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            print(f"{amount} is withdraw.New balacne is {self.balance}")
    
## create an account
account=BankAccount("krish",5000)
print(account.balance)
account.deposite(300)
account.withdraw(500)

        
```
