```python
#Padndas DataFrame and Sries
#Series:- A pandas series is one-dimensional arrays-like object that can hold any data type.It is similar to a column in a table
import pandas as pd
data=[1,2,3,4,5]
series_pd=pd.Series(data)
print(series_pd)
```

```python
#create a Series form dictonary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
## Dataframe

## create a dataframe from a duictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Banglore','New York','Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```
