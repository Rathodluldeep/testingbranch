Seaborn
Seaborn is a python visualization library based on Matplotlib that provides a high-level uinterfcae for drawing attractive and informative statiscal graphics.seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Basic ploting with seaborn
tips=sns.load_dataset("tips")
tips
```

```python
## scatter polot 

sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title('Scatter plot of total bill vs tip')
plt.show()
```

```python
##line plot

sns.lineplot(y="total_bill",x="size",data=tips)
plt.title("Line plot of total bill vs tip")
plt.show()
```

```python
## catagerial plot
sns.catplot(x="day",y="total_bill",data=tips)
plt.title("Catagerial plot of total bill vs day")
plt.show()
```

```python
## bar plot
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("Bar plot of total bill vs day")
plt.show()
```

```python
## box plot
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("Box plot of total bill vs day")  
plt.show()
```

```python
## violin plot
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("Violin plot of total bill vs day")
plt.show()
```

```python
## histogram plot
sns.histplot(x="total_bill",data=tips,bins=10,kde=True)
plt.title("Histogram plot of total bill")
plt.show()
```

```python
##kde plot
sns.kdeplot(x="total_bill",data=tips,fill=True)
plt.title("KDE plot of total bill") 
plt.show()
```

```python
## Pairplot
sns.pairplot(tips)
```

```python
#heatmap
corr=sns.heatmap(tips[['total_bill','tip','size']].corr(),annot=True)
corr
```
