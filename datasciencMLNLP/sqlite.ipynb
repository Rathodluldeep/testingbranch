SQL and SQLite

SQl(Structured Query Launguage) is a standard launguage for managing and manupulating relational databases.SDLite is a self-contained,serverless,and zero-configuration database engine that is widely used for embedded database systems.In this lesson,we will cover the basics of SQL and SQLite,including creating databases,tables,and performing various SQL operations.

```python
import sqlite3
```

```python
##Coonect to the SQLite database
connections=sqlite3.connect('example.db')
c=connections.cursor()
```

```python
## create a table
c.execute('''Create Table If Not Exists employees(
    ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Age INTEGER,
    department TEXT
)''')

## commit the chnages to the database
connections.commit()
```

```python
c.execute('''
          Select * from employees
''')
```

```python
## Insert the data into the table
c.execute('''
          Insert into employees(Name,Age,department) values('John Doe',30,'IT')
''')
c.execute('''
          Insert into employees(Name,Age,department) values('Bob',25,'Engineering')
''')
c.execute('''
          Insert into employees(Name,Age,department) values('Charlie',35,'Finance')
''')

```

```python
##Querry the data from the table
c.execute('''
          Select * from employees  
''')
rows=c.fetchall()

for row in rows:
    print(row)
## close the connection
```

```python
## Update the data in the table

c.execute('''Update employees set Age=31 where Name='John Doe' ''')
connections.commit()
##Querry the data from the table
c.execute('''       Select * from employees  ''')
rows=c.fetchall()   
for row in rows:
    print(row)
```

```python
## Delete the data from the table
c.execute('''Delete from employees where Name='John Doe' ''')
connections.commit()
##Querry the data from the table
c.execute('''       Select * from employees  ''')
rows=c.fetchall()
for row in rows:
    print(row)  
```

```python
## working with sales data
c.execute('''Create Table If Not Exists sales(
    ID INTEGER PRIMARY KEY,
    Product TEXT NOT NULL,
    Date TEXT NOT NULL,
    sales Integer,
    region INTEGER
)''')

## Insert the data into the table
sales_data=[
    ('Product 1','2023-01-01',100,'North'),
    ('Product 2','2023-01-02',200,'South'),
    ('Product 3','2023-01-03',150,'East'),
    ('Product 4','2023-01-04',300,'West'),
    ('Product 5','2023-01-05',250,'North')   
]

c.executemany('''Insert into sales(Product,Date,sales,region) values(?,?,?,?)''',sales_data)
connections.commit()
```

```python
##Query data from the sales table
c.execute('select * from sales')
sales_rows=c.fetchall()

#Print the sales data
for row in sales_rows:
    print(row)
```

```python
## close the connection
connections.close()
```
